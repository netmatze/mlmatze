```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
array2D = np.random.random((10,10))
```

```python
from IPython.display import clear_output
import time

x = array2D[0]
y = array2D[1]

m = -0.1
b = 0.1
m_n = 0.005
b_n = 0.0001
learning_rate_m=0.01
learning_rate_b=0.001
epochs=80
# def linear_regression(x, y, m = -0.1, b = 0.1, learning_rate=0.01, epochs=20):

fig, axs = plt.subplots(8, 10, figsize=(30, 40))
row = 0
column = 0
y_pred = [0 for i in range(1,10+1)]
for epoch in range(epochs):
  error = 0
  # clear_output(wait=True)
  for index, value in enumerate(x):
    y_pred[index] = m * value + b
    # add substract learning rate
    # if value > y_pred[index]:
    #   m = m + m_n
    #   b = b + b_n
    # elif value < y_pred[index]:
    #   m = m - m_n
    #   b = b - b_n
    # add learning rate times y - y_pred
    m = m + learning_rate_m * (y[index] - y_pred[index])
    b = b + learning_rate_b * (y[index] - y_pred[index])
    # add learning rate times y - y_pred or substract learning rate times y - y_pred
    # if value > y_pred[index]:
    #   m = m + learning_rate_m * (y[index] - y_pred[index])
    #   b = b + learning_rate_b #* (y[index] - y_pred[index])
    # elif value < y_pred[index]:
    #   m = m - learning_rate_m * (y[index] - y_pred[index])
    #   b = b - learning_rate_b #* (y[index] + y_pred[index])
  for i, y_i in enumerate(y):
    error = error + abs(y_i - y_pred[i])
  axs[column, row].plot(x, m*x+b, color = 'red')
  axs[column, row].title.set_text(str(error))
  axs[column, row].scatter(x, y_pred, color = 'green')
  axs[column, row].scatter(x, y)
  plt.xlabel('Independent variable X')
  plt.ylabel('Dependent variable y')
  row = row + 1
  if row > 9:
    row = 0
    column = column + 1
  # time.sleep(1)
```
